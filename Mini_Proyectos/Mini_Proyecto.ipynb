## Introduccin
El estudio de vibraciones en rotores es esencial en el diseo de mquinas rotatorias,
ya que fenmenos como la inestabilidad dinmica, el **whirl** (precesin hacia adelante o atrs) y
las resonancias pueden comprometer el desempeo y la seguridad de equipos industriales.
En particular, los efectos de **no linealidad en curvatura e inercia** producen desviaciones
respecto al comportamiento lineal, que deben modelarse con ecuaciones diferenciales no lineales.

El artculo *Free vibrations analysis of a rotating shaft with nonlinearities in curvature and inertia*
(Hosseini & Khadem, 2009) presenta un modelo continuo basado en la teora de Euler-Bernoulli,
posteriormente reducido mediante el mtodo de Galerkin. El resultado es un sistema dinmico
acoplado en dos direcciones transversales, con trminos cbicos caractersticos de los osciladores
tipo Duffing.

---

## Descripcin del fenmeno
El eje rotatorio se deforma lateralmente en dos planos ortogonales, descritos por las coordenadas
generalizadas V(t) y W(t). La rotacin introduce:
- **acoplamiento giroscpico**, que relaciona las velocidades transversales,
- **no linealidades cbicas** en la rigidez efectiva, tanto individuales (V, W) como acopladas (VW, WV),
- **efectos de amortiguamiento** que limitan la amplitud de las vibraciones.

Este fenmeno se traduce en trayectorias complejas en el espacio de estados, que pueden incluir
oscilaciones estables, inestables y patrones de precesin.

---

## Modelo analtico y anlisis cualitativo
El modelo reducido no lineal es:

$$ M \ddot{V} + c \dot{V} + K V + G \Omega \dot{W} + \gamma V^3 + \eta V W^2 = 0 $$
$$ M \ddot{W} + c \dot{W} + K W - G \Omega \dot{V} + \gamma W^3 + \eta W V^2 = 0 $$

donde:
- **V, W**: desplazamientos transversales,
- **M**: masa modal efectiva,
- **c**: amortiguamiento modal,
- **K**: rigidez modal,
- **G**: coeficiente giroscpico,
- ****: velocidad angular de giro,
- ****: coeficiente cbico de rigidez no lineal,
- ****: coeficiente de acoplamiento cbico.

Segn la no-dimensionalizacin y la discretizacin del artculo:

- M = 1 + nI    masa modal efectiva,
- K = n           rigidez modal,
- G = 2nI      coeficiente giroscpico,
-   n           coeficiente cbico principal,
-                    coeficientes mixtos cbicos,

El anlisis cualitativo se centra en la bsqueda de equilibrios (soluciones estacionarias)

**Modelo reducido**

El modelo matemtico simplificado que representa las vibraciones laterales de un eje rotatorio con
no linealidades en curvatura e inercia es un sistema acoplado tipo Duffing:

$$ M \ddot{V} + c \dot{V} + K V + G \Omega \dot{W} + \gamma V^3 + \eta V W^2 = 0 $$
$$ M \ddot{W} + c \dot{W} + K W - G \Omega \dot{V} + \gamma W^3 + \eta W V^2 = 0 $$

donde:
- **V, W**: desplazamientos laterales del eje en dos planos ortogonales (x e y).
- **V_d, W_d**: velocidades laterales.
- **V_dd, W_dd**: aceleraciones laterales.

---

**Significado de los parmetros fsicos**

- **M**: masa modal efectiva (adimensional). Incluye la inercia del eje y el efecto giroscpico
  del giro.
- **c**: coeficiente de amortiguamiento viscoso. Modela la disipacin de energa por friccin.
- **K**: rigidez modal. Representa la resistencia del eje a deformarse lateralmente.
- **G**: coeficiente giroscpico. Est ligado a la inercia rotacional transversal *I*.
- ** (Omega)**: velocidad angular de giro del eje. Aparece en los trminos giroscpicos
  que acoplan V y W, dando lugar a fenmenos de *whirl* (precesin hacia adelante o hacia atrs).
- ** (gamma)**: coeficiente de rigidez no lineal cbica (tipo Duffing). Surge de la curvatura
  no lineal del eje.
- ** (eta)**: coeficiente de acoplamiento no lineal. Modela la interaccin cbica entre las dos
  direcciones de desplazamiento (trminos mixtos VW y WV).

---

**Segn la no-dimensionalizacin y la discretizacin del artculo:**

- M = 1 + nI   masa modal efectiva.
- K = n         rigidez modal.
- G = 2nI    coeficiente giroscpico en la forma lineal usada.
-   n         coeficiente cbico principal que produce la dureza no lineal.
-                  agrupa los coeficientes de los trminos mixtos cbicos
(de orden similar a  en la escala).
- c                 amortiguamiento modal.
-                  velocidad de giro del eje.

```python
# Dependencias
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from ipywidgets import interact, FloatSlider, IntSlider, fixed, Checkbox
import sympy as sp
```

## Definicin del sistema (funcin RHS)
Definimos la funcin que devuelve las derivadas del sistema de 1er orden.


```python
def system_rhs(t, y, params):
    """
    y = [V, Vdot, W, Wdot]
    params: dict con claves M, c, K, G, Omega, gamma, eta
    """
    V, Vd, W, Wd = y
    M = params['M']
    c = params['c']
    K = params['K']
    G = params['G']
    Omega = params['Omega']
    gamma = params['gamma']
    eta = params['eta']

    # Ecuaciones (segn signo y convencin):
    Vdd = (-c*Vd - K*V - G*Omega*Wd - gamma*(V**3) - eta*V*(W**2)) / M
    Wdd = (-c*Wd - K*W + G*Omega*Vd - gamma*(W**3) - eta*W*(V**2)) / M

    return [Vd, Vdd, Wd, Wdd]


# ## Parmetros por defecto (puedes cambiarlos en la celda interactiva)

# %%
def default_params(n=1, I2=6.25e-4, c=0.05, Omega=10.0, eta_factor=0.2):
    """
    Construye parmetros M,K,G,gamma,eta a partir de los parmetros modalizados usados en el artculo.
    n: nmero de modo
    I2: momento diametral adimensional
    eta_factor: fraccin del coeficiente cbico principal para generar el trmino mixto
    """
    pi = np.pi
    M = 1.0 + (n**2)*(pi**2)*I2
    K = (n**4)*(pi**4)
    G = 2.0*(n**2)*(pi**2)*I2
    gamma = (n**6)*(pi**6)
    eta = eta_factor * gamma
    return {'M': M, 'c': c, 'K': K, 'G': G, 'Omega': Omega, 'gamma': gamma, 'eta': eta}


# ## 1) Anlisis cualitativo: encontrar equilibrios y estabilidad (analtico/simblico)
# Buscaremos puntos de equilibrio (V*, W*, Vd*=0, Wd*=0) resolviendo las ecuaciones algebraicas.

def find_equilibria_symbolic(params):
    # Definimos variables simblicas
    V, W = sp.symbols('V W', real=True)
    K = params['K']
    gamma = params['gamma']
    eta = params['eta']

    eq1 = sp.Eq(K*V + gamma*V**3 + eta*V*W**2, 0)
    eq2 = sp.Eq(K*W + gamma*W**3 + eta*W*V**2, 0)

    sols = sp.solve([eq1, eq2], [V, W], dict=True)
    return sols

# %%
# Jacobiano del sistema en la forma de primer orden (simblico)
V, Vd, W, Wd = sp.symbols('V Vd W Wd')
M_s, c_s, K_s, G_s, Omega_s, gamma_s, eta_s = sp.symbols('M c K G Omega gamma eta')

# Variables de estado: x = [V, Vd, W, Wd]
# Formulamos las RHS simblicamente para poder derivar el jacobiano
Vdd_s = (-c_s*Vd - K_s*V - G_s*Omega_s*Wd - gamma_s*V**3 - eta_s*V*W**2) / M_s
Wdd_s = (-c_s*Wd - K_s*W + G_s*Omega_s*Vd - gamma_s*W**3 - eta_s*W*V**2) / M_s

f1 = Vd
f2 = Vdd_s
f3 = Wd
f4 = Wdd_s

X = sp.Matrix([f1, f2, f3, f4])
states = sp.Matrix([V, Vd, W, Wd])

Jac = X.jacobian(states)


# Funcin que evala la estabilidad lineal en un punto de equilibrio (calcula autovalores del jacobiano).

# %%
def linear_stability_at_equilibrium(eq_point, params):
    # eq_point: dict {'V': val, 'W': val}
    subs = {V: eq_point['V'], W: eq_point['W'], Vd: 0, Wd: 0,
            M_s: params['M'], c_s: params['c'], K_s: params['K'],
            G_s: params['G'], Omega_s: params['Omega'],
            gamma_s: params['gamma'], eta_s: params['eta']}
    J_num = np.array(Jac.subs(subs).evalf(), dtype=float)
    eigvals, eigvecs = np.linalg.eig(J_num)
    return eigvals, J_num


# ## 2) Simulaciones de PVI con solve_ivp
# Funcin que integra y devuelve la solucin

# %%
def simulate_ivp(y0, t_span, params, rtol=1e-8, atol=1e-10, method='RK45'):
    sol = solve_ivp(lambda t,y: system_rhs(t,y,params), t_span, y0,
                    method=method, rtol=rtol, atol=atol, dense_output=True)
    return sol


# ## Funciones de graficado

# %%
def plot_time_series(sol, vars_to_plot=('V','W'), figsize=(10,4)):
    t = sol.t
    y = sol.y
    plt.figure(figsize=figsize)
    if 'V' in vars_to_plot:
        plt.plot(t, y[0,:], label='V (desplazamiento)')
    if 'Vdot' in vars_to_plot:
        plt.plot(t, y[1,:], label='V_dot')
    if 'W' in vars_to_plot:
        plt.plot(t, y[2,:], label='W (desplazamiento)')
    if 'Wdot' in vars_to_plot:
        plt.plot(t, y[3,:], label='W_dot')
    plt.xlabel('t')
    plt.legend()
    plt.grid(True)
    plt.show()

# %%
def plot_phase_space(sol, kind='V_Vdot'):
    y = sol.y
    plt.figure(figsize=(5,5))
    if kind == 'V_Vdot':
        plt.plot(y[0,:], y[1,:])
        plt.xlabel('V')
        plt.ylabel('V_dot')
    elif kind == 'W_Wdot':
        plt.plot(y[2,:], y[3,:])
        plt.xlabel('W')
        plt.ylabel('W_dot')
    elif kind == 'V_W':
        plt.plot(y[0,:], y[2,:])
        plt.xlabel('V')
        plt.ylabel('W')
    plt.grid(True)
    plt.show()


# ## 3) Ejemplo: anlisis analtico y simulacin con parmetros por defecto

# %%
# Parmetros por defecto
params = default_params(n=1, I2=6.25e-4, c=0.05, Omega=10.0, eta_factor=0.1)
print('Parmetros usados (ejemplo):')
for k,v in params.items():
    print(f"  {k} = {v}")

# Encontrar equilibrios simblicamente (puede devolver soluciones evidentes como (0,0))
sols = find_equilibria_symbolic(params)
print('\nEquilibrios simblicos encontrados:')
for s in sols:
    print(s)

# Estabilidad lineal en el equilibrio trivial (0,0)
evals, Jnum = linear_stability_at_equilibrium({'V':0,'W':0}, params)
print('\nJacobiano linealizado en (0,0):')
print(Jnum)
print('\nAutovalores en (0,0):')
print(evals)




# ## 4) Celda interactiva: correr simulaciones cambiando parmetros y condiciones iniciales
# Usa los sliders para variar condiciones y ver resultados. Esta celda llama a `solve_ivp` cada vez que
# cambias un parmetro del `interact`.

# %%
@interact(
    V0=FloatSlider(min=-0.05, max=0.05, step=0.001, value=0.01, description='V(0)'),
    Vd0=FloatSlider(min=-1.0, max=1.0, step=0.01, value=0.0, description="V'(0)"),
    W0=FloatSlider(min=-0.05, max=0.05, step=0.001, value=0.0, description='W(0)'),
    Wd0=FloatSlider(min=-1.0, max=1.0, step=0.01, value=0.0, description="W'(0)"),
    Omega=FloatSlider(min=0.0, max=100.0, step=1.0, value=10.0, description='Omega'),
    c=FloatSlider(min=0.0, max=1.0, step=0.01, value=0.05, description='c'),
    I2=FloatSlider(min=0.0, max=0.01, step=1e-4, value=6.25e-4, description='I2'),
    eta_factor=FloatSlider(min=0.0, max=1.0, step=0.01, value=0.1, description='eta_fac'),
    tmax=FloatSlider(min=1.0, max=200.0, step=1.0, value=100.0, description='tmax'),
    plot_time=Checkbox(value=True, description='Mostrar serie temporal'),
    plot_phase=Checkbox(value=True, description='Mostrar espacio de estados')
)
def run_interactive(V0, Vd0, W0, Wd0, Omega, c, I2, eta_factor, tmax, plot_time, plot_phase):
    params = default_params(n=1, I2=I2, c=c, Omega=Omega, eta_factor=eta_factor)
    y0 = [V0, Vd0, W0, Wd0]
    t_span = (0.0, float(tmax))
    sol = simulate_ivp(y0, t_span, params, method='RK45')

    if plot_time:
        plot_time_series(sol, vars_to_plot=('V','W','Vdot','Wdot'))
    if plot_phase:
        plot_phase_space(sol, kind='V_Vdot')
        plot_phase_space(sol, kind='W_Wdot')
        plot_phase_space(sol, kind='V_W')
```

Observaciones rpidas:
- Si todos los autovalores tienen parte real negativa, el equilibrio es linealmente estable.
- En presencia de giro (Omega) los trminos giroscpicos pueden desplazar modos y producir acoplamientos
  que generan oscilaciones con crecimiento o batido segn otros parmetros.
